We again assume that our words in plaintext form are made up
of the signs „0“ and „1“, i.e. the plaintext alphabet is A = {0, 1}.
Like in the example, the ciphertext alphabet is B = {v, w, x, y, z}.
In common texts the „0“ occurs with a probability of 60 %, and
the „1“ with a probability of 40 %. 

Furthermore, the bigrams „00“,
„01“, „10“ respectively „11“ occur with a probability of 32 %, 29 %,
29 % respectively 10 %.

0 60%

1 40%

========================


00 32%

01 29%

10 29%

11 10%

============================

```
def singlegram(text):
  F = {}
  for c in text:
    if c in F:
      F[c]+=1
    else:
      F[c]=1
  print (F)
  total=0
  for e in F:
    total+=F[e]
  print (total)
  for e in F:
    print ("element :" + e + ", percentage: " + str(F[e]/total*100))

def bigram(text):
  F={}
  key=""
  for i in range(len(text)-1):
    key=text[i]+text[i+1]
    if key in F:
      F[key]+=1
    else:
      F[key]=1
  print (F)
  total=0
  for e in F:
    total+=F[e]
  print (total)
  for e in F:
    print ("element :" + e + ", percentage: " + str(F[e]/total*100))
```

```
singlegram (cipher)
bigram(cipher)
```

```
# solution (vwxyz = 001)
S=['1','1','0','0','0']
clear=cipher
clear=clear.replace("v",S[0])
clear=clear.replace("w",S[1])
clear=clear.replace("x",S[2])
clear=clear.replace("y",S[3])
clear=clear.replace("z",S[4])

singlegram(clear)
bigram (clear)
print (clear)
```
