```
#example test vectors


KEY="VAUX"
CIPHER="VUNLKYRQSTFGZ"

ALPHABET="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```
def decrypt(cipher,key):
  
  plain=""
  KEY=key
  i=0
  for c in cipher:
    plain+=chr((ALPHABET.index(cipher[i])-ALPHABET.index(KEY[i]))%26 + 65)
    KEY=key+plain
    i=i+1
  return (plain)  
  
```

```
print(decrypt(CIPHER,KEY))
```

```
import random
def randomkey():
  l = list(ALPHABET)
  random.shuffle(l)
  result = ''.join(l[0:5])
  return (result)
```
