```
#example

# read strips from url

import requests

url = 'https://www.mysterytwisterc3.org/images/challenges/mtc3-schmeh-09-stripset.txt'

r = requests.post(url)

#print (r.text)

```

# New Section

```
S = []

for line in r.text.splitlines():
    S.append(line[3:])

print (S[66])
```

```
KEY = [66, 11, 52,55, 4, 90]
OFFSET=11

cipher="UKLAGW"
plain="CRYPTO"



def encrypt():
  l=0
  encrypted_text=""
  for c in plain:
    currentstrip=S[KEY[l]]
    encrypted_text+=currentstrip[(currentstrip.index(c)+OFFSET)%26]
    l+=1
  print (encrypted_text)
  
def decrypt():
  l=0
  decrypted_text=""
  for c in cipher:
    
    currentstrip=S[KEY[l]]
    decrypted_text+=currentstrip[(currentstrip.index(c)-OFFSET)%26]
    l+=1
  print (decrypted_text)  
  
```
