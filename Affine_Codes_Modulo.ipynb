```
# a ciphertext letter y is: y = a  x + b mod 26

cipher="GNXBNBSXNUXFXWXNDXUHPWKNPWS"
alphabet="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
codewheel=""

a=17
b=7

for i in range(26):
  coded=(i*a+b)%26
  codewheel+=alphabet[coded]
  
print (codewheel)  
```

```
plain=""

for i in range(len(cipher)):
  plain+=alphabet[codewheel.index(cipher[i])]

print (plain)
```
