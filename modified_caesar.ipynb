```
def decrypt (alphabet,line,shift):
    sol=""
    #print ("decrypting "+ line + " shift: " + str(shift) )
    for c in line:
      pos=alphabet.index(c)
      sol+=alphabet[(pos+shift)%52]
    return sol      
```
