```
def chi_square_german(text):
  
  GDIST={
      "A": 5.58,"B": 1.96,"C": 3.16,"D": 4.98,"E": 16.93,"F": 1.49,"G": 3.02,"H": 4.98,"I": 8.02,"J": 0.24,"K": 1.32,"L": 3.60,"M": 2.55,
      "N": 10.53,"O": 2.24,"P": 0.67,"Q": 0.02,"R": 6.89,"S": 6.42,"T": 5.79,"U": 3.83,"V": 0.84,"W": 1.78,"X": 0.05,"Y": 0.05,"Z": 1.21,
  }
  
  ALPHABET="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  COUNTS= [0]*26
  CHISQUARE= [0]*26
  #text=text.replace("X","")
  #text=text.replace("Y","")
  for c in text:
    COUNTS[ALPHABET.index(c)]+=1
  #print("text len: " + str(len(text)))
  for curchar in ALPHABET:
    i=ALPHABET.index(curchar)
    expected=GDIST[curchar]/100*len(text)
    #print (expected)
    CHISQUARE[i]= ((COUNTS[i]-expected)* (COUNTS[i]-expected))/expected
    
    
  print (COUNTS[23])
  #print (CHISQUARE)
  return sum(CHISQUARE)
  
      
  
```
