```
def singlegram(text):
  F = {}
  for c in text:
    if c in F:
      F[c]+=1
    else:
      F[c]=1
  print (F)
  total=0
  for e in F:
    total+=F[e]
  print (total)
  for e in F:
    print ("element :" + e + ", percentage: " + str(F[e]/total*100))

def bigram(text):
  F={}
  key=""
  for i in range(len(text)-1):
    key=text[i]+text[i+1]
    if key in F:
      F[key]+=1
    else:
      F[key]=1
  print (F)
  total=0
  for e in F:
    total+=F[e]
  print (total)
  for e in F:
    print ("element :" + e + ", percentage: " + str(F[e]/total*100))
```

```
#singlegram(cipher)
#bigram (cipher)

# solution (xyz = 001)
S=['0','0','1']
clear=cipher
clear=clear.replace("x",S[0])
clear=clear.replace("y",S[1])
clear=clear.replace("z",S[2])

singlegram(clear)
bigram (clear)
print (clear)
```
