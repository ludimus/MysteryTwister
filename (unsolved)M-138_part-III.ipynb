```
# helper functions find overlapping substrings

def find_substring(line,substring):
  lss=len(substring)
  count=0
  
  for i in range(len(line)-lss+1):
    #print ("comparing:" + line[i:i+lss] + " "  + substring)
    if line[i:i+lss]==substring:
      count+=1
  return(count)    

#print (str(find_substring("tralala","ala")))
```
